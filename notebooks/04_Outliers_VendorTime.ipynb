```python
import pandas as pd

df = pd.read_parquet('../data/clean_yellow_tripdata_2025-01.parquet')
```

```python
# Identify extreme trips
outliers = df[df['trip_distance'] > 50]
print(f"Number of extreme trips: {len(outliers)}")
```

```python
df = df[df['trip_distance'] <= 50]
```

```python
df['long_trip'] = df['trip_distance'] > 10
```

```python
vendor_kpis_clean = df.groupby('VendorID').agg(
    total_revenue=('total_amount','sum'),
    avg_revenue=('total_amount','mean'),
    num_trips=('total_amount','count'),
    avg_tip_pct=('tip_pct','mean'),
    long_trip_pct=('long_trip','mean'),
    avg_distance=('trip_distance','mean')
).reset_index()

vendor_kpis_clean['long_trip_pct'] *= 100
vendor_kpis_clean.to_csv('../data/vendor_kpis_clean.csv', index=False)
```

```python
time_kpis = df.groupby('pickup_hour').agg(
    total_revenue=('total_amount','sum'),
    avg_revenue=('total_amount','mean'),
    num_trips=('total_amount','count'),
    avg_distance=('trip_distance','mean'),
    avg_tip_pct=('tip_pct','mean')
).reset_index()

time_kpis.to_csv('../data/time_kpis_clean.csv', index=False)
```

```python
df.to_parquet('../data/clean_yellow_tripdata_2025-01.parquet', index=False)
df.to_csv('../data/clean_yellow_tripdata_2025-01.csv', index=False)
```
