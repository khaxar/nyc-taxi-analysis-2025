```python
# What hours generate the most revenue?
df['pickup_hour'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.hour
revenue_by_hour = df.groupby('pickup_hour')['total_amount'].sum()
revenue_by_hour.sort_values(ascending=False)
revenue_by_hour.plot(kind='bar')
```

```python
# Which zones have the longest trips?
avg_distance = df.groupby('PULocationID')['trip_distance'].mean()
avg_distance.sort_values(ascending=False).head(10)
```

```python
# Are tips correlated with distance?
df[['trip_distance', 'tip_amount']].corr()
```

```python
df.sample(50000).plot.scatter(
    x='trip_distance',
    y='tip_amount',
    alpha = 0.2,
)
```

```python
# Are weekends different from weekdays?
df['pickup_date'] = pd.to_datetime(df['tpep_pickup_datetime']).dt.date
df['weekday'] = pd.to_datetime(df['pickup_date']).dt.weekday
df['is_weekend'] = df['weekday'] >= 5
df.groupby('is_weekend')['total_amount'].mean()
```

```python
# Which vendors underperform?
df.groupby('VendorID')['total_amount'].mean()
```
