```python
import pandas as pd

df = pd.read_parquet('../data/clean_yellow_tripdata_2025-01.parquet')
locations = pd.read_csv('../data/taxi_zone_lookup.csv')

df = df.merge(locations, left_on='PULocationID', right_on='LocationID', how='left')
```

```python
locations.head()
# df.head()
```

```python
# Hourly Demand & Revenue

hourly = df.groupby('pickup_hour').agg(
    total_revenue=('total_amount', "sum"),
    num_trips=('total_amount', 'count'),
    avg_revenue=('total_amount', 'mean'),
    avg_tip=('tip_pct', 'mean')
).reset_index()

hourly.to_csv('D:/analyst/projects/nyc-taxi-analysis-2025/data/business_hourly_demand.csv', index=False)
```

```python
# high-revenue but low-trip zones

zone_demand = df.groupby(['Borough','Zone']).agg(
    total_revenue=('total_amount','sum'),
    avg_revenue=('total_amount','mean'),
    num_trips=('total_amount','count')
).reset_index()

zone_demand.to_csv('D:/analyst/projects/nyc-taxi-analysis-2025/data/business_zone_demand.csv', index=False)
```

```python
# Vendor efficiency

df['revenue_per_mile'] = df['total_amount'] / df['trip_distance']

vendor_eff = df.groupby('VendorID').agg(
    total_revenue=('total_amount','sum'),
    avg_revenue=('total_amount','mean'),
    avg_distance=('trip_distance','mean'),
    revenue_per_mile=('revenue_per_mile','mean'),
    avg_tip=('tip_pct','mean')
).reset_index()

vendor_eff.to_csv('D:/analyst/projects/nyc-taxi-analysis-2025/data/business_vendor_efficiency.csv', index=False)
```

```python
# Long trip value

long_trip_analysis = df.groupby('long_trip').agg(
    avg_revenue=('total_amount','mean'),
    avg_duration=('trip_duration_min','mean'),
    avg_distance=('trip_distance','mean'),
    avg_tip=('tip_pct','mean')
).reset_index()

long_trip_analysis.to_csv('D:/analyst/projects/nyc-taxi-analysis-2025/data/business_long_trip_value.csv', index=False)
```
